```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.simplefilter('ignore')
#plt.style.use('dark_background')
```

```python
data = pd.read_csv('Transformed_Housing_Data2.csv')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler as scl
```

```python
scaler = scl()
```

```python
Y = data['Sale_Price']
X = scaler.fit_transform(data.drop(columns = ['Sale_Price']))
X = pd.DataFrame(data = X, columns = data.drop(columns = ['Sale_Price']).columns)
```

```python
X.head()
```

# Checking and removing Multicollinearity

```python
X.corr()
```

```python
k = X.corr()
z = [[str(i),str(j)] for i in k.columns for j in k.columns if (k.loc[i,j]>abs(0.5))&(i!=j)]#loops (simple and nested) in simple opersnd type
# A single line comprehension
z, len(z)
```

# Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
vif_data = X
VIF = pd.Series([vif(vif_data.values,i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
# Variable for which the VIF is maximum
VIF[VIF== VIF.max()].index[0]
```

```python
def MC_remover(data):
    VIF = pd.Series([vif(data.values,i) for i in range(data.shape[1])], index = data.columns)
    if VIF.max() > 5:
        column = VIF[ VIF == VIF.max()].index[0]
        print(column, 'has been removed')
        data = data.drop(columns = column)
        return data
    else:
        print('No Multicollinearity present')
        return data
```

```python
for i in range(7):
    vif_data= MC_remover(vif_data)

vif_data.head()
```

# Remaining Columns

```python
VIF = pd.Series([vif(vif_data.values,i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF, len(vif_data.columns)
```

# Train/Test Set

```python
x = vif_data
y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.3, random_state = 101)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression as LR
lr = LR(normalize = True)
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
# Generating Predictions over test set
predictions =  lr.predict(x_test)

# Finally calulating the score (R2 value) for the predicted set
lr.score(x_test, y_test)
```

# Hence Completed
Considering this is a Linear Regression Model 84% accuracy is considered quite good.

# Verifying Assumptions
1. Residuals

```python
residuals = predictions - y_test
residual_table = pd.DataFrame({'residuals': residuals, 
                              'predictions': predictions})

residual_table = residual_table.sort_values(by = 'predictions')
```

```python
z = [i for i in range(int(residual_table['predictions'].max()))]
k = [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi = 130, figsize = (17,7))
plt.scatter(residual_table['predictions'], residual_table['residuals'], color = 'red', s = 2)
plt.plot(z, k, color = 'green', linewidth = 3, label = 'Regression Line')
plt.ylim(-800000, 800000)
plt.xlabel('Fitted Points (Ordered by Predictions)')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.legend()
plt.show()
```

# Distribution of Errors

```python
plt.figure(dpi = 100, figsize = (10,7))
plt.hist(residual_table['residuals'], color = 'red', bins= 200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('Distribution of Residuals')
plt.show()
```

Interpreting the Results
