```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings(action= 'ignore')
```

```python
data = pd.read_csv('data3')
```

```python
data['Zipcode']= data['Zipcode'].astype(object)
```

```python
pd.set_option('max_columns',None)
data.head(10)
```

```python
data["No of Times Visited"].unique()
```

```python
mapping = {'None':'0','Once':'1','Twice':'2','Thrice':'3','Four':'4'}
```

```python
data['No of Times Visited'] = data['No of Times Visited'].map(mapping)
```

```python
data['No of Times Visited'].head(10)
```

```python
data['Ever Renovated'] = np.where(data['Renovated Year']== 0,'No','Yes')
```

```python
data['Year Renovated'] = pd.DatetimeIndex(data['Date House was Sold']).year
```

```python
data['Years since Renovation']= np.where(data['Ever Renovated']== 'Yes', abs(data['Year Renovated'] - data['Renovated Year']),0)
```

```python
data.head(10)
```

```python
data.drop(columns = ['Year Renovated','Renovated Year','Date House was Sold'], inplace = True)
```

```python
data.to_csv('Transformed_data')
```
